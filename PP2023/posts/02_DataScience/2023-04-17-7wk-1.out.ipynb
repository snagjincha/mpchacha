## 이미지자료의 이해 및 시각화

`-` 참고: [빛의삼원색](https://ko.wikipedia.org/wiki/%EC%9B%90%EC%83%89)

`-` 이미지자료의 이해: 이미지자료는 rgb값이 저장된 3개의 매트릭스를
쌓은것이라 이해가능

-   이미지자료의 shape은 (가로픽셀, 세로픽셀, 3) 이다. // 5\*5
    이미지라면 (5,5,3)
-   3은 RGB 빛의 밝기를 정의함.
-   밝기는 0-255 사이의 int, 0-1 사이의 float으로 정한다.
-   int 255 혹은 float 1 은 해당색이 매우 밝다는 것을 의미함.
-   각 픽셀의 색깔은 RGB의 밝기 (RGB의 값) 으로 조정가능하다. (따라서 총
    256*256*256 개의 색상을 표현할 수 있다.)

`-` 예시1

```python
r = np.array([0]*25*3).reshape(5,5,3) 
g = np.array([0]*25*3).reshape(5,5,3) 
b = np.array([0]*25*3).reshape(5,5,3) 
```

```python
r[:3,:3,0] = 255   
```

```python
r[:3,:3,0] = 255   
g[:3,2:,1] = 255
b[2:,:,2] = 255 
```

```python
r[:,:,0], r[:,:,1], r[:,:,2]
```

```python
g[:,:,0], g[:,:,1], g[:,:,2]
```

```python
b[:,:,0], b[:,:,1], b[:,:,2]
```

```python
plt.imshow(r+g+b)
```

`-` 예시2: r,g,b에 작은 숫자를 주면 점점 어두워 진다.

```python
r = np.array([0]*25*3).reshape(5,5,3) 
g = np.array([0]*25*3).reshape(5,5,3) 
b = np.array([0]*25*3).reshape(5,5,3) 
```

```python
r[:3,:3,0] = 80
g[:3,2:,1] = 80
b[2:,:,2] = 80
```

```python
plt.imshow((r+g+b))
```

`2-3`. 아래와 같은 합성함수를 생각하라.

-   $l(x)=w_0+w_1x$
-   $a(x)=\frac{1}{1+e^{-x}}$

`2`. $x \in [-5,5]$의 범위에서 $a(l(x))$의 그래프를 그려라. 단 이때
$w_0=0, w_1=1$로 설정한다.

(풀이)

$y=a(l(x))= a(w_0+w_1x) = 1/(1+\exp(-w_0-w_1x))$

그런데, $w_0=0$ 이고 $w_1=1$ 이므로 $y=a(l(x)) = 1/ (1+\exp(-x))$ 이다.

```python
x = np.linspace(-5,5,100)
y = 1/(1+np.exp(-x))
plt.plot(x,y)
```

`3`. 아래는 $a(l(x))$의 성질에 대하여 토의한 내용이다. 올바르게 서술한
학생을 골라라.

-   하영: $w_0=0,w_1=1$ 이면 $a(l(x))$는 증가하는 함수이다.
-   재인:
-   서연:
-   보람:
-   지윤:

(풀이)

아래와 같은 함수를 선언한뒤에 하나씩 체크해 나가면된다.

```python
def l(x,w0=0,w1=1): 
    return w0+w1*x 
def a(x): 
    return 1/(1+np.exp(-x))
```

남겨서 “빨간하니”를 만들고 이미지를 출력하라.

```python
red_hani = hani*0 
red_hani[:,:,0] = hani[:,:,0]
plt.imshow(red_hani)
```

`4`. 하니의 모든 값에 루트를 취하여 “루트하니”를 만들고 “원본하니”와
“루트하니”를 좌우로 나란히 배치하여 출력하라.

(풀이)

```python
root_hani = np.sqrt(hani)
plt.imshow(np.concatenate([hani,root_hani],axis=1))
```

“루트하니”의 이미지는 “원본하니”의 이미지와 비교하여 어떤가? 왜 그러한
결과가 나왔다고 생각하는가?

`5`. 하니의 모든값에 아래와 같은 함수를 적용하라.

-   $f(x)=\begin{cases} \sqrt{x} & x>0.7 \\ x & x \leq 0.7 \end{cases}$

함수의 결과로 얻어진 매트릭스를 “후광하니”라고 부르자. “원본하니”와
“후광하니”를 좌우로 나란히 배치하여 출력하라.

```python
shiny_hani = np.sqrt(hani)*(hani>0.7)+hani*(hani<=0.7)
plt.imshow(np.concatenate([hani,shiny_hani],axis=1))
```
